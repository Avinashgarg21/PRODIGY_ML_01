### It can be observed that the test data have salePrice missing and that is exactly what i need to predict here, so ulltimately our task is to create a csv file that has the the ID and their corrosponding salePrice for all the test data.

```python
# Selecting relevent features

features = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
```

```python
# Fill missing values with the median of each column, for both train and test

train_df[features] = train_df[features].fillna(train_df[features].median())

test_df[features] = test_df[features].fillna(test_df[features].median())
```

```python
# Now we need to extract features and target variable from training data

X_train = train_df[features]

y_train = train_df['SalePrice']
```

```python
# Now extract features from test data

X_test = test_df[features]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize and train the model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict the sale price for the test data

test_df['SalePrice'] = model.predict(X_test)
```

```python
# Creating a submission DataFrame to store the sale price of the test data corrosponding to ID of each house

submission_df = test_df[['Id', 'SalePrice']]
submission_df.to_csv('submission.csv', index=False)
```

```python
# Visualization of linear regression line
plt.figure(figsize=(10, 6))
```

```python
plt.scatter(train_df['GrLivArea'], y_train, color='black', label='Training Data')

gr_liv_area_range = np.linspace(train_df['GrLivArea'].min(), train_df['GrLivArea'].max(), 500)

median_values = train_df[features].median()
predict_df = pd.DataFrame({
    'GrLivArea': gr_liv_area_range,
    'BedroomAbvGr': median_values['BedroomAbvGr'],
    'FullBath': median_values['FullBath']
})

predicted_prices = model.predict(predict_df)

plt.plot(gr_liv_area_range, predicted_prices, color='red', linewidth=2, label='Regression Line')

plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('Linear Regression: GrLivArea vs SalePrice')
plt.legend()

plt.show()
```
